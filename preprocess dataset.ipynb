```python
import pandas as pd
```

```python
electric_cars_raw = pd.read_csv("https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD")
electric_cars_raw
```

```python
electric_cars = electric_cars_raw \
    [["City", "County", "Model Year", "Make", "Model", "Electric Vehicle Type"]] \
    .value_counts() \
    .reset_index() \
    .rename(columns={0: "n_cars"})
electric_cars.columns = electric_cars.columns \
    .str.lower() \
    .str.replace(" ", "_") \
    .values
electric_cars
```

```python
electric_cars.to_csv("electric_cars.csv", index=False)
```
